```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

def initialize_webdriver():
    options = webdriver.ChromeOptions()
    #options.add_argument('--headless')
    options.add_experimental_option("prefs", {
      "download.default_directory": r'C:\Users\iida.lassila\Downloads',
      "download.prompt_for_download": False,
      "download.directory_upgrade": True,
      "safebrowsing.enabled": True
    })
    driver = webdriver.Chrome(options=options)
    return driver


def download_wine_price_list():

    driver = initialize_webdriver()

    url = "https://www.alko.fi/valikoimat-ja-hinnasto/hinnasto"
    driver.get(url)
    
    time.sleep(10)
    
    try:
        download_button = driver.find_element(By.XPATH, '//*[@id="js-main-wrap"]/div[2]/div[2]/div[1]/div[2]/div[3]/div/div/div/div[2]/div/div/div/div[2]/div[2]/a')
        download_button.click()

        time.sleep(10)

    except Exception as e:
        print(f"Exception: {e}")
        
def read_wine_price_excel(file_path):
    try:
        df = pd.read_excel(file_path, header=3)
        return df
    except Exception as e:
        print(f"Exception: {e}")
        return None

    
download_wine_price_list()
file_name = r'C:\Users\iida.lassila\Downloads\alkon-hinnasto-tekstitiedostona.xlsx'
df = read_wine_price_excel(file_name)
df
```

```python
df.rename(columns={
    'Numero': 'ID',
    'Nimi': 'Name',
    'Valmistaja': 'Producer',
    'Pullokoko': 'Size',
    'Hinta': 'Price',
    'Valmistusmaa': 'Country',
    'Vuosikerta': 'Year',
    'Rypäleet': 'Grapes',
    'Alkoholi-%': 'Alcohol-%',
    'Sokeri g/l': 'Sugar g/l',
    'Energia kcal/100 ml': 'kcal/100 ml'
}, inplace=True)

df = df[['ID', 'Name', 'Producer', 'Size', 'Price', 'Country', 'Year', 'Grapes', 'Alcohol-%', 'Sugar g/l', 'kcal/100 ml']]
df
```
